# Predict  a students final exam score based on the nuber of hours they study Using Linear Regression

#### Step 1 :

```python

import numpy as np
import  pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.model_selection import  train_test_split
from sklearn.linear_model import LinearRegression 
```

#### step 2 :

```python
data ={ "Hours_study" : [2,2.5,2.8,3,3.5,3.8,4,4.5,4.8,5,6,7,8,9,10,11],
       "Exam_score" : [50,55,56,60,65,68,70,77,79,79,81,86,90,91,92,98]
      }
df= pd.DataFrame(data)
```

```python
df.head()
```

```python
plt.scatter(df["Hours_study"] ,df["Exam_score"])  
plt.xlabel("Hours_study") 
plt.ylabel("Exam_score")
```

#### step 4 :

```python
x= df[["Hours_study"]]
y=df[["Exam_score"]] 
```

```python
y
```

#### step 5 :


```python
x_train, x_test , y_train , y_test = train_test_split(x,y,test_size=0.2 ,random_state=42) 
```

```python
x_train
```

#### step 6:

```python
model = LinearRegression()
```

#### step 7 :

```python
model.fit(x_train,y_train)
```

# User Input Testing : 

```python
user_input = float(input("Enter the number of hours you study :")) 

predicted_score = model.predict([[user_input]]) 
print("Predicted Exam score :",predicted_score[0][0])
```
